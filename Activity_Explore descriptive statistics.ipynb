AQI  is Air Quality index a tool that measures, reports daily air quality
What it does
The AQI is a scale that measures air quality from 0 to 500, with 0 being perfect air and 500 being an immediate danger to public health. The AQI also indicates the health effects that may occur within hours or days of breathing polluted air. 
How it works
The AQI is based on measurements of pollutants like ground-level ozone, sulfur dioxide, nitrogen dioxide, carbon monoxide, and fine particles. The AQI breaks air pollution levels into six categories, each with a color and advice. 
What it means
AQI values below 100 are considered satisfactory, while values above 100 indicate unhealthy air quality. The higher the AQI value, the greater the risk of health harm. 
Who developed it
The U.S. Environmental Protection Agency (EPA) developed the AQI to provide a simple, consistent way to report air quality
